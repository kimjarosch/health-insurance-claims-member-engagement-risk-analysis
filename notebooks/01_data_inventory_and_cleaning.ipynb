```python
# data quality check
def quick_profile(df, name):
    print(f"\n{name}")
    print("-" * 60)
    print("Shape:", df.shape)
    print("\nTop 10 missing-value counts:")
    print(df.isnull().sum().sort_values(ascending=False).head(10))
    print("\nDuplicate full rows:", df.duplicated().sum())

quick_profile(benef_2009, "Beneficiary 2009")
quick_profile(benef_2010, "Beneficiary 2010")
quick_profile(inp, "Inpatient Claims")
quick_profile(outp, "Outpatient Claims")
```

```python
# keep columns that matter
benef_2009_small = benef_2009[beneficiary_keep].copy()
benef_2010_small = benef_2010[beneficiary_keep].copy()
inp_small = inp[inpatient_keep].copy()
outp_small = outp[outpatient_keep].copy()

print("benef_2009_small:", benef_2009_small.shape)
print("benef_2010_small:", benef_2010_small.shape)
print("inp_small:", inp_small.shape)
print("outp_small:", outp_small.shape)
```

```python
# convert data columns
benef_date_cols = ["BENE_BIRTH_DT", "BENE_DEATH_DT"]
inp_date_cols = ["CLM_FROM_DT", "CLM_THRU_DT", "CLM_ADMSN_DT", "NCH_BENE_DSCHRG_DT"]
outp_date_cols = ["CLM_FROM_DT", "CLM_THRU_DT"]

for col in benef_date_cols:
    benef_2009_small[col] = pd.to_datetime(benef_2009_small[col], errors="coerce")
    benef_2010_small[col] = pd.to_datetime(benef_2010_small[col], errors="coerce")

for col in inp_date_cols:
    inp_small[col] = pd.to_datetime(inp_small[col], errors="coerce")

for col in outp_date_cols:
    outp_small[col] = pd.to_datetime(outp_small[col], errors="coerce")
```
